## Build Funnel Chart Using Area Chart Structure

## 1. Total Bill Counts


### Include Y axis labels for only the top row (to reduce clutter in the chart)

```python
funnel1 = alt.Chart(df[df.YEAR < 2014]).mark_area().encode(
x=alt.Y("TOTAL", axis=None).stack('center'),
y=alt.X("STATUS:N", axis=alt.Axis(domain=False, title="", offset=15, labelFontSize=15, 
                    labelFont='Arial', ticks=False, grid=False), sort='ascending'),
column=alt.Column("YEAR", title='', spacing=0, header=alt.Header(labelFontSize=15, labelPadding=-10))
).properties(
width=75, 
height=150,
title=alt.TitleParams(
        text='Arizona State Legislature',
        subtitle='A look at legislation trends within each legislative session in the last 15 years',
        fontSize=20,
        subtitlePadding=6,
        color='black',
        anchor='middle',
    dx=60,
    dy=-20
    ))

funnel1
```

```python
funnel2 = alt.Chart(df[(df.YEAR < 2019) & (df.YEAR > 2013)]).mark_area().encode(
x=alt.Y("TOTAL", axis=None).stack('center'),
y=alt.X("STATUS:N", axis=alt.Axis(domain=False, labels=False, title="", ticks=False, grid=False), sort='ascending'),
column=alt.Column("YEAR", title='', spacing=0, header=alt.Header(labelFontSize=15, labelPadding=-10))
).properties(
width=75, height=150)

funnel2
```

```python
funnel3 = alt.Chart(df[(df.YEAR > 2018)]).mark_area().encode(
x=alt.Y("TOTAL", axis=None).stack('center'),
y=alt.X("STATUS:N", axis=alt.Axis(domain=False, labels=False, title="", ticks=False, grid=False), sort='ascending'),
# color="YEAR:N", 
    column=alt.Column("YEAR", title='', spacing=0, header=alt.Header(labelFontSize=15, labelPadding=-10))
).properties(
width=75, height=150)

funnel3
```

### Combine all plots

```python
alt.vconcat(funnel1, funnel2, funnel3, spacing=-200).configure_view(
stroke=None).display(
    scaleFactor=3
)
```

## 2. Plot with House & Senate Detail Using Color

```python
# transform into long data format
df_chamber = df.melt(id_vars=['STATUS', 'YEAR'], value_name='BILL_COUNT', var_name='CHAMBER')
```

```python
# remove unnecessary rows
df_chamber = df_chamber[df_chamber.CHAMBER != 'TOTAL']
```

```python
df_chamber.head()
```

```python
funnel1 = alt.Chart(df_chamber[df_chamber.YEAR < 2014]).mark_area().encode(
x=alt.Y("BILL_COUNT", axis=None).stack('center'),
y=alt.X("STATUS:N", axis=alt.Axis(domain=False, title="", offset=15, labelFontSize=15, 
                    labelFont='Arial', ticks=False, grid=False), sort='ascending'),
color="CHAMBER", 
column=alt.Column("YEAR", title='', spacing=0, header=alt.Header(labelFontSize=15, labelPadding=-12))
).properties(
width=85, 
height=150,
title=alt.TitleParams(
        text='Arizona State Legislature',
        subtitle='A look at legislation trends within each legislative session in the last 15 years',
        subtitlePadding=6,
        fontSize=20,
        color='black',
        anchor='middle',
    dx=60,
    dy=-20
    ))

funnel1
```

```python
funnel2 = alt.Chart(df_chamber[(df_chamber.YEAR < 2019) & (df_chamber.YEAR > 2013)]).mark_area().encode(
x=alt.Y("BILL_COUNT", axis=None).stack('center'),
y=alt.X("STATUS:N", axis=alt.Axis(domain=False, labels=False, title="", ticks=False, grid=False), sort='ascending'),
color="CHAMBER", 
column=alt.Column("YEAR", title='', spacing=0, header=alt.Header(labelFontSize=15, labelPadding=-12))
).properties(
width=85, height=150)

funnel2
```

```python
funnel3 = alt.Chart(df_chamber[(df_chamber.YEAR > 2018)]).mark_area().encode(
x=alt.Y("BILL_COUNT", axis=None).stack('center'),
y=alt.X("STATUS:N", axis=alt.Axis(domain=False, labels=False, title="", ticks=False, grid=False), sort='ascending'),
color="CHAMBER", 
column=alt.Column("YEAR", title='', spacing=0, header=alt.Header(labelFontSize=15, labelPadding=-12))
).properties(
width=85, height=150)

funnel3
```

```python
alt.vconcat(funnel1, funnel2, funnel3, spacing=-200).configure_view(
stroke=None).configure_legend(
labelFontSize=12,
titleFontSize=14
).display(
    scaleFactor=3
)
```
