## Olympic Medal Analysis - July 2024 in anticipation of Paris Olympics

Focusing on gold medal winners, which events have been most consistently dominated, and by what countries?

dataset https://www.kaggle.com/datasets/piterfm/olympic-games-medals-19862018/data

```python
# import libraries

import altair as alt
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read in data
df = pd.read_csv('olympics/olympic_medals.csv')
```

```python
# extract year
df['year'] = df.slug_game.str.slice(-4).astype(int)
```

```python
df.shape
```

```python
df.year.value_counts().sort_index(ascending=False).plot();
```

```python
# only include gold medals
df = df[df.medal_type == 'GOLD']
```

```python
df.shape
```

## Clean up "event_title" column

```python
def event_title_clean_up(row):
    event_title = str(row['event_title'])
    event_gender = str(row['event_gender'])
    if event_gender == event_title:
        return event_gender
    elif event_gender in event_title:
        return event_title.replace("Men's", '').replace("Women's", '').replace("Mixed", '')
    elif event_gender.lower() in event_title:
        return event_title.replace(event_gender.lower(), '')
    else:
        return event_title
```

```python
df.event_title = df.apply(event_title_clean_up, axis=1)
```

```python
df.event_title = df.event_title.str.replace(' x ', 'x')
```

```python
df.event_title = df.event_title.str.lower()
```

```python
df.event_title = df.event_title.str.strip()
```

```python
df.head()
```

### For events with gender duplication between 'event title' and 'event_gender'

```python
amb_event_title = df[df.discipline_title.isin(df[df.event_title.isin(['men', 'women'])].discipline_title)]
```

```python
amb_event_title.discipline_title.value_counts()
```

```python
amb_event_title.event_title = ''
```

### Non-abmiguous 'event title' and 'event_gender'

```python
event_title = df[~df.discipline_title.isin(df[df.event_title.isin(['men', 'women'])].discipline_title)]
```

## Identify the events events with the highest run counts

```python
all_events = pd.concat([amb_event_title,
                        event_title])
```

```python
all_events.event_title = all_events.event_title.str.lower()
```

```python
all_events.head()
```

```python
all_events.shape
```

## Find streaks

```python
# returns length of the longest list
def find_sequences_separated_by_four(lst):
    result = []
    sequence = []
    longest_length = 0
    for i in range(len(lst)-1):
        if abs(lst[i+1] - lst[i]) == 4:
            sequence.append(lst[i])
        elif sequence:
            sequence.append(lst[i])  # Include the last number in the sequence
            result.append(sequence)
            longest_length = max(longest_length, len(sequence))
            sequence = []
    if sequence:
        sequence.append(lst[-1])  # Include the last number in the sequence
        result.append(sequence)
        longest_length = max(longest_length, len(sequence))
    return longest_length
```

```python
count_diff_four_years = all_events.groupby(['discipline_title', 'event_title', 'event_gender', 'country_name'])['year'].unique().apply(find_sequences_separated_by_four)
```

```python
count_diff_four_years.value_counts()
```

```python
count_diff_four_years
```

```python
# include only greater than specific length
top_run_events = count_diff_four_years[count_diff_four_years > 4].index
```

```python
top_streaks_all_years = all_events.set_index(['discipline_title', 'event_title', 'event_gender', 'country_name']).loc[top_run_events].reset_index()
```

```python
top_streaks_all_years.shape
```

## For events with the longest streaks, extract years and format data into long format

```python
# returns the longest list
def find_sequences_separated_by_four_list(lst):
    result = []
    sequence = []
    longest_sequence = []
    for i in range(len(lst)-1):
        if abs(lst[i+1] - lst[i]) == 4:
            sequence.append(lst[i])
        elif sequence:
            sequence.append(lst[i])  # Include the last number in the sequence
            result.append(sequence)
            if len(sequence) > len(longest_sequence):
                longest_sequence = sequence
            sequence = []
    if sequence:
        sequence.append(lst[-1])  # Include the last number in the sequence
        result.append(sequence)
        if len(sequence) > len(longest_sequence):
            longest_sequence = sequence
    return longest_sequence
```

```python
top_streaks_all_years = top_streaks_all_years.groupby(['discipline_title', 'event_title', 'event_gender', 'country_name'])['year'].unique().apply(find_sequences_separated_by_four_list)
```

```python
top_streaks_long = top_streaks_all_years.apply(pd.Series).melt(ignore_index=False, value_name='year').drop(columns='variable').reset_index().dropna()
```

```python
top_streaks_long
```

# Final clean up before plotting

```python
# format as datatime
top_streaks_long.year = pd.to_datetime(top_streaks_long.year, format='%Y')
```

```python
# create detailed name 
top_streaks_long['discipline_event_title_gender'] = top_streaks_long['discipline_title'] + ": " + top_streaks_long['event_title'] + ' - ' + top_streaks_long['event_gender']
```

```python
top_streaks_long.shape
```

```python
# add column for length of the streak
top_streaks_long = top_streaks_long.merge(top_streaks_long.groupby(['country_name', 'discipline_event_title_gender']).size().rename('gold_medal_streak_length').reset_index())
```

```python
top_streaks_long.gold_medal_streak_length.value_counts()
```

```python
# exclude streak that is interferring with the correct sorting of the longer People's Republic of China streak

top_streaks_long.loc[(top_streaks_long.discipline_event_title_gender == 'Diving: 3m springboard - Women'), 'gold_medal_streak_length'] = 9
```

```python
# create counter column to be used to create horizontal count for each event
top_streaks_long['year_count'] = 1
```

```python
top_streaks_long = top_streaks_long.sort_values(['country_name', 'discipline_event_title_gender', 'year'])
```

```python
top_streaks_long['year_count'] = top_streaks_long.groupby(['discipline_event_title_gender', 'country_name'])['year_count'].cumsum()
```

```python
# add additional year for continuing streaks
top_streaks_long = pd.concat([top_streaks_long,
           top_streaks_long.loc[top_streaks_long.year == top_streaks_long.year.max()].assign(year=pd.to_datetime('2024-01-01'))])
```

```python
top_streaks_long.loc[top_streaks_long.year == top_streaks_long.year.max(), 'continuing'] = True
```

```python
top_streaks_long.loc[top_streaks_long.continuing.notna(), 'year_count'] = ' '
```

```python
top_streaks_long.loc[top_streaks_long.continuing.notna()]
```

# Plot

```python
text = alt.Chart(top_streaks_long).mark_text(size=8, color='white', dx=-.5, dy=1, fontWeight='bold').encode(
        x=alt.X("year:T"),
        y=alt.Y("discipline_event_title_gender", sort=alt.EncodingSortField(field="gold_medal_streak_length", order='descending')),
        text='year_count:O'
)
# text
```

```python
arrow = alt.Chart(top_streaks_long).mark_point(filled=True, fillOpacity=1).encode(
        x=alt.X("year:T"),
        y=alt.Y("discipline_event_title_gender", sort=alt.EncodingSortField(field="gold_medal_streak_length", order='descending')),
        angle=alt.AngleValue(90),
        size=alt.SizeValue(200),
        shape=alt.condition(
            alt.datum.continuing == True,
            alt.ShapeValue(value="triangle"),        
            alt.ShapeValue(value='None')),
        color='country_name',
opacity=alt.value(1))

# arrow
```

```python
lines = alt.Chart(top_streaks_long).mark_rule(strokeCap='round').encode(
    x=alt.X("max(year)", title=None, scale=alt.Scale(domain=["1896-01-01", "2020-01-01"], type="time")),
    y=alt.Y("discipline_event_title_gender", 
            title=None, 
            axis=alt.Axis(labelLimit=300, domain=False, labelPadding=10, grid=True, gridDash=[3,5], ticks=False),
            sort=alt.EncodingSortField(field="gold_medal_streak_length", order='descending'),
           ),
    color=alt.Color("country_name", title="Country Name"),
    x2=alt.Y2("min(year)"),
    strokeWidth=alt.value(3),
    
)
# lines
```

```python
circles = alt.Chart(top_streaks_long).mark_point(opacity=1).encode(
    x=alt.X("year", scale=alt.Scale(domain=["1896-01-01", "2024-01-01"], type="time")),
    y=alt.Y("discipline_event_title_gender", sort=alt.EncodingSortField(field="gold_medal_streak_length", order='descending')),
    color=alt.Color("country_name", title=None, ),
    strokeWidth=alt.value(8),
    shape=alt.condition(
        alt.datum.continuing == True,
        alt.ShapeValue(value="None"),        
        alt.ShapeValue(value='circle')),
)

# circles
```

```python
title=alt.Title(
       "Top Consecutive Gold Medal 🥇 Streaks",
    offset=40, 
    fontSize=25
)
```

```python
(lines + circles + arrow + text).properties(
    width=600,
    height=700,
    padding=20,
    title=title  # Set the title offset to 20 pixels
).configure_view( # turn off border around the plot
    stroke=None
).configure_legend(
    title=None,
    symbolOffset=15, 
#     labelOffset=40,
    labelFontSize=12,
    orient='none',
    labelBaseline='middle',
    offset=25,
    direction='vertical',
    legendY = 250,
    legendX = 600
).configure_axis(
    labelFontSize=13  # Set the fontsize for both x and y axis labels
).display(
    scaleFactor=4
)
```
