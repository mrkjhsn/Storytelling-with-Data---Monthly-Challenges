## Import data (downloaded manually from this website)

https://www.flightradar24.com/data/airports/phx/arrivals

```python
arrival_data1 = pd.read_clipboard()
```

```python
arrival_data1.shape
```

```python
arrival_data2 = pd.read_clipboard()
```

```python
arrival_data2.shape
```

```python
arrival_data3 = pd.read_clipboard()
```

```python
arrival_data3.shape
```

```python
arrival_data = pd.concat([arrival_data1,
                          arrival_data2,
                          arrival_data3])
```

```python
arrival_data.shape
```

```python
arrival_data = arrival_data.reset_index(drop=True)
```

```python
# arrival_data.to_json()
```

```python
# filter for only landed aircraft
arrival_data = arrival_data[arrival_data.STATUS.str.contains('Landed')]
```

```python
arrival_data.shape
```

```python
arrival_data.AIRCRAFT = arrival_data.AIRCRAFT.str.split('(').str[0].str.strip()
```

```python
arrival_data
```

## Consolidate aircraft names since there is some ambiguity in sub models of aircraft

```python
arrival_data.AIRCRAFT = arrival_data.AIRCRAFT.replace(['B38M', 'B738', 'B739', 'B39M'], 'B737')

arrival_data.AIRCRAFT = arrival_data.AIRCRAFT.replace(['A21N', 'A320', 'A20N'], 'A321')
```

```python
arrival_data.AIRCRAFT.value_counts().head(10)
```

```python
arrival_data.AIRCRAFT.value_counts().head(10).index

# .sort_index()
```

```python
arrival_data_top_10 = arrival_data[arrival_data.AIRCRAFT.isin(arrival_data.AIRCRAFT.value_counts().head(10).index)]
```

```python
arrival_data_top_10
```

```python
arrival_data_top_10['ARRIVAL'] = arrival_data_top_10.STATUS.str[-8:]
```

```python
arrival_data_top_10.ARRIVAL = pd.to_datetime("2024-07-08 " + arrival_data_top_10.ARRIVAL)
```

```python
arrival_data_top_10
```

## Aggregate aircraft count per hour of the day

```python
arrival_agg = arrival_data_top_10.groupby([arrival_data_top_10.ARRIVAL.dt.hour, 'AIRCRAFT']).size().rename('COUNT').reset_index()
```

```python
arrival_agg['xOffset'] = random.choices(range(-15, 16), k=arrival_agg.shape[0])
```

```python
arrival_agg['ANGLE'] = random.choices(range(-3, 10), k=arrival_agg.shape[0])
```

```python
arrival_agg
```

```python
arrival_agg['ARRIVAL'] = pd.to_datetime("2024-07-08 " + arrival_agg.ARRIVAL.astype(str) + ':00')
```

```python
aircraft_sort = ['PC12',
                 'E135',
                 'E75L',
                 'A319',
                 'A321',
                 'B737',
                 'CRJ7',
                 'B763',
                 'B752',
                 'C208']
```

```python
arrival_agg['xOffset'] = random.choices(range(-15, 16), k=arrival_agg.shape[0])
```

```python
# create title and subtitle
title=alt.Title(
       'Phoenix Sky Harbor - Arrival Flights',
       subtitle="Aggregated by aircraft type and arrival time (5 day period)"
)


alt.Chart(arrival_agg).mark_point(
    filled=True, opacity=1
).encode(
    alt.X("AIRCRAFT", sort=aircraft_sort, title='Aircraft Type', axis=alt.Axis(domain=False, labelAngle=360, titleFontSize=15, labelPadding=10,
                                                                               labelFontSize=15, tickSize=14, tickBand='center', tickOffset=0,
                                                                              titlePadding=18)),
    alt.Y("ARRIVAL:T", timeUnit='hoursminutes', title='Time of Day', 
          axis=alt.Axis(domain=False, titleFontSize=15, grid=True, gridDash=[3,5])),
    alt.ShapeValue(airplane52),
    size=alt.Size('COUNT', scale=alt.Scale(range=[0, 40])),
#     angle='ANGLE',
    xOffset='xOffset',  
    color=alt.Color('AIRCRAFT', legend=None)
).properties(
    title=title,
    width=900,
    height=800
).configure_title(
    font='Arial',
    fontSize=18,
).configure_legend(
    title=None,
    symbolOffset=40, 
    labelOffset=40,
    labelFontSize=14).display(
    scaleFactor=2
)
```
